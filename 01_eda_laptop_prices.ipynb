# Laptop Price Data Analysis

# Dataset Overview

A subset of the Laptop Price Prediction using Specifications dataset was used for educational purposes.

Source: Kaggle | Laptop Price Prediction using Specifications

# Dataset Parameters
The dataset includes the following attributes:

Manufacturer: Laptop manufacturer

Category: Type of laptop (mapped to values 1–5)

GPU: GPU manufacturer (mapped to values 1–3)

OS: Operating system (Windows = 1, Linux = 2)

CPU_core: Processor type (mapped to i3, i5, i7 as 3, 5, 7)

Screen_Size_cm: Screen size in centimeters

CPU_frequency: Processor frequency (GHz)

RAM_GB: RAM size in gigabytes

Storage_GB_SSD: SSD storage size (GB)

Weight_kg: Laptop weight in kilograms

Price: Price in USD
