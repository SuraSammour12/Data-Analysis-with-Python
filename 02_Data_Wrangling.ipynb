# Deal with missing data

How should i deal with missing data?

Drop data
* a. Drop the whole row
* b. Drop the whole column

Replace data
* a. Replace it by mean
* b. Replace it by frequency
* c. Replace it based on other functions



---
| **Method**                    | **When to Use It**                                                                | **Suitable Column Type** |
| ----------------------------- | --------------------------------------------------------------------------------- | ------------------------ |
| **Replace by mean**           | When the column is numeric and has few missing values                             | Numeric                  |
| **Replace by mode/frequency** | When the column is categorical and there's a dominant frequent value              | Categorical              |
| **Drop the row**              | When the missing value cannot be reasonably replaced, or it's the target variable | Any type                 |

```python
# Calculate the mean value for the "normalized-losses" column
avg_norm_loss = df['normalized-losses'].astype('float').mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
# Replace "NaN" with mean value in "normalized-losses" column
df['normalized-losses']=df['normalized-losses'].fillna(avg_norm_loss)
```

```python
# Calculate the mean value for the "bore" column
avg_bore = df['bore'].astype('float').mean(axis=0)
```

```python
# Replace "NaN" with the mean value in the "bore" column
df['bore']=df['bore'].fillna(avg_bore)
```

```python
# Replace NaN in "stroke" column with the mean value
avg_stroke = df['stroke'].astype('float').mean(axis=0)
df['stroke']=df['stroke'].fillna(avg_stroke)
```

```python
# Replace "NaN" with the mean value in the "horsepower" column
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
df['horsepower']=df['horsepower'].fillna(avg_horsepower)
```

```python
# Replace "NaN" with the mean value in the "peak-rpm" column
avg_peakrpm = df['peak-rpm'].astype('float').mean(axis=0)
df['peak-rpm'] = df['peak-rpm'].fillna(avg_peakrpm)
```

```python
# To see which values are present in a particular column, i can use the ".value_counts()" method
df['num-of-doors'].value_counts()
```

i can see that four doors is the most common type.

```python
# i can also use the ".idxmax()" method to calculate the most common type automatically:
df['num-of-doors'].value_counts().idxmax()
```

```python
#replace the missing 'num-of-doors' values by the most frequent
df["num-of-doors"] = df['num-of-doors'].fillna('four')
```

Finally, drop all rows that do not have price data:

```python
df.dropna(subset=['price'],axis=0,inplace=True)
```

# Data Standardization
Standardization is the process of transforming data into a common format, allowing the researcher to make the meaningful comparison.

In the current dataset, the fuel consumption in the columns "city-mpg" and "highway-mpg" is expressed in mpg (miles per gallon). Since the application being developed targets a country that uses the liters per 100 kilometers (L/100km) standard, it was necessary to transform the data accordingly.

The values were converted from mpg to L/100km using the formula:

L/100km = 235 / mpg

This transformation allows for fuel consumption to be interpreted in a format aligned with international standards, particularly in countries that use the metric system, making it easier to compare vehicle performance.

```python
df.head()
```

```python
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-mpg'] = 235/df['city-mpg']
# rename column name from "city-mpg" to "city-L/100km"
df.rename(columns={'city-mpg':'city-L/100km'},inplace=True)
df.head()
```

```python
df["highway-mpg"] = 235 / df["highway-mpg"]

df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)

df.head()
```

# Data Normalization

Normalization is the process of transforming values of several variables into a similar range.

```python
df.columns
```

When working with variables such as "length", "width", and "height", it can be helpful to bring their values to a common scale to simplify comparison and analysis. In this case, normalization is applied so that the values of each variable range from 0 to 1.

This is done by dividing each value in the column by the maximum value of that same column, following the formula:

normalized_value = (original value) / (maximum value)

As a result, the largest value becomes 1, and all other values fall proportionally between 0 and 1.



```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
print(df[['length','width','height']])
```

# Binning
Binning is a process of transforming continuous numerical variables into discrete categorical 'bins' for grouped analysis.


```python
# horsepower column
df['horsepower'] = df['horsepower'].astype('float')
print(df['horsepower'].min())
print(df['horsepower'].max())
print(df['horsepower'].nunique())
```

In the dataset, the "horsepower" column contains real numerical values ranging from 48 to 262, with 59 unique values in total. However, when the goal is to compare car prices based on general horsepower categories rather than precise numbers, it becomes more practical to group the values into broader categories.

To simplify the analysis, the horsepower values can be divided into three distinct bins, representing low, medium, and high horsepower ranges. This transformation allows for a clearer understanding of trends or differences in price across general horsepower levels, instead of analyzing every specific value.

The cut function from the Pandas library can be used to perform this segmentation, turning continuous numerical data into categorical bins.

```python
# Plot the histogram of horsepower to see the distribution of horsepower
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df['horsepower'])

plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('count')
plt.pyplot.title('horsepower bins')
```

To create three bins of equal width from the "horsepower" column, the `numpy.linspace()` function can be used. The starting point is set to the minimum value of the "horsepower" column, and the ending point is set to its maximum value. Since the goal is to divide the range into three equal parts, four evenly spaced numbers are needed to represent the bin edges. These four values define the boundaries of the three bins.


```python
bins = np.linspace(min(df['horsepower']), max(df['horsepower']),4)
bins
```

```python
# set group names
group_names = ['Low','Medium','High']
```

Apply the function `'cut'` to determine what each value of df['horsepower'] belongs to.

```python
df['horsepower-binned'] = pd.cut(df['horsepower'],bins,labels=group_names,include_lowest=True)
df[['horsepower','horsepower-binned']].head(20)
```

See the number of vehicles in each bin:

```python
df['horsepower-binned'].value_counts()
```

Plot the distribution of each bin:

```python
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

I successfully reduced the 59 unique horsepower values into just 3 intervals.

# Indicator Variable

What is an indicator variable?

An indicator variable (or dummy variable) is a numerical variable used to label categories. They are called 'dummies' because the numbers themselves don't have inherent meaning.

Why use indicator variables?

Indicator variables are used to convert categorical variables into a numeric form, enabling their use in regression analysis in later modules.

Example

The column "fuel-type" has two unique values: "gas" or "diesel". Regression doesn't understand words, only numbers. To use this attribute in regression analysis, i can convert "fuel-type" to indicator variables.

Use the Panda method `'get_dummies'` to assign numerical values to different categories of fuel type.

```python
df.columns
```

```python
dummy_variable_1 = pd.get_dummies(df['fuel-type'])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

In the data frame, column 'fuel-type' now has values for 'gas' and 'diesel' as 0s and 1s.

merge data frame 'df' and 'dummy_variable_1'


```python
df = pd.concat([df,dummy_variable_1],axis=1)
```

drop original column 'fuel-type' from 'df'

```python
df.drop('fuel-type',axis=1,inplace=True)
```

```python
df.head()
```

Similar to before, create an indicator variable for the column "aspiration"

```python
dummy_variable_2 = pd.get_dummies(df['aspiration'])
dummy_variable_2.head()
```

```python
dummy_variable_2.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)
```

```python
dummy_variable_2.head()
```

Merge the new dataframe to the original dataframe, then drop the column 'aspiration'.

```python
df = pd.concat([df,dummy_variable_2],axis=1)
```

```python
df.drop(['aspiration'],axis=1,inplace=True)
```
