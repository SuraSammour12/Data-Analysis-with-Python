# Kernel Density Estimation (KDE) Plots for Model Evaluation

1. What is a KDE Plot?

A Kernel Density Estimation (KDE) Plot is a method to visualize the probability density function (PDF) of a dataset. It provides a smooth curve representing the distribution of data points.

2. Why Use KDE Plots in Model Evaluation?

KDE plots are smoother and less sensitive to bin size than histograms.

They help in comparing actual values vs. predicted values in regression models.

KDE plots can reveal deviations and misfits in model predictions.

They are effective for visual inspection of model performance.

# Example Workflow - Regression Model Evaluation:

```python
# Import Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Generate synthetic data (X and Y with noise)
np.random.seed(42)
x = np.random.rand(100)*10
y = 3*x+np.random.normal(0,3,100) # y=3x+Noise | Linear relation with noise
data = pd.DataFrame({'x':x, 'y':y})
```

```python
# Split the data into training and testing sets
x_train,x_test,y_train,y_test = train_test_split(data[['x']], data[['y']], test_size=0.2, random_state=42) # Training Set 80% | Testing Set 20%
```

```python
# Train a Linear Regression model
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
# Predict using the trained model
y_pred = model.predict(x_test)
```

```python
# Plot the KDE plots for actual vs. predicted values

plt.figure(figsize=(8, 5))
sns.kdeplot(y_test.squeeze(), label='Actual', color='blue', linewidth=2)
sns.kdeplot(y_pred.squeeze(), label='Predicted', color='red', linewidth=2)
plt.xlabel('Target Variable')
plt.ylabel('Density')
plt.title('KDE Plot of Actual vs. Predicted Values')
plt.legend()
plt.show()
```

The regression model successfully captures the general shape of the data distribution.

It underestimates the variance and fails to model the distribution's extremes.

The model’s performance could be improved by using more flexible techniques like Polynomial Regression or Regularization to better fit the data’s complexity.
