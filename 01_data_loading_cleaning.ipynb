#  Introduction to Dataset Analysis with Pandas

## 1. Data Acquisition

- Datasets can be stored locally or online, and in formats like `.csv`, `.json`, or `.xlsx`.
- We used a CSV file of automobile data hosted online.
- To read CSV data into a Pandas DataFrame:
```python

````python
import pandas as pd
import numpy as np
````

````python
file_path='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv'
````

````python
df = pd.read_csv(file_path)
````

# 2. Inspecting the Data
Use df.head(n) to see the top n rows.

Use df.tail(n) to view the last n rows.

````python
df.tail(5)
````

# 3. Adding Column Headers

````python
headers = [
    "symboling", "normalized-losses", "make", "fuel-type", "aspiration",
    "num-of-doors", "body-style", "drive-wheels", "engine-location", "wheel-base",
    "length", "width", "height", "curb-weight", "engine-type", "num-of-cylinders",
    "engine-size", "fuel-system", "bore", "stroke", "compression-ratio", "horsepower",
    "peak-rpm", "city-mpg", "highway-mpg", "price"
]
df.columns = headers
````

 Column headers are assigned based on the official dataset documentation

 Source: https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.names


````python
df.head()
````



---

# 4. Data Cleaning

````python
df=df.replace('?',np.nan)
````

````python
df
````

````python
# Remove rows with missing price values:
df=df.dropna(subset=['price'], axis=0)
````

````python
df
````

# 5. Exploring the Data

````python
# Check data types
df.dtypes
````

````python
# Get statistical summaries - Numeric only
df.describe()
````

````python
# Get statistical summaries - All columns
df.describe(include='all')
````

For object columns, .describe returns:

- `unique`: Number of unique values
- `top`: Most frequent value
- `freq`: Count of the top value

````python
# Use df.info() to display structure, data types, and NON-null counts.
df.info()
````

# 6. Save Processed Data


````python
# To save the cleaned dataset:
df.to_csv('automobile.csv',index=False)
````
