# 1. Linear Regression and Multiple Linear Regression

# Simple Linear Regression
How just one variable can help us predict price.

For example, 'highway-mpg' as an influence on 'price'

**General equation:**
Yhat = a+bx

* a refers to the intercept of the regression line, in other words: the value of Y when X is 0

* b refers to the slope of the regression line, in other words: the value with which Y changes when X increases by 1 unit


```python
# load the modules for linear regression:
from sklearn.linear_model import LinearRegression
```

```python
# Create the linear regression object:
lm = LinearRegression()
lm
```

How could 'highway-mpg' help us predict car price?

Using simple linear regression, i will create a linear function with 'highway-mpg' as the predictor variable and the 'price' as the response variable

```python
X = df[['highway-mpg']]
Y = df['price']
```

```python
# Fit the linear model using highway-mpg:
lm.fit(X,Y)
```

```python
# Output a prediction:
Yhat = lm.predict(X)
Yhat[0:5]
```

```python
# Intercept (a)
print(lm.intercept_)
```

```python
# Slope (b)
print(lm.coef_)
```

# What is the final linear model we get?
Yhat = a + bX

Plugging in the actual values we get:

Price = 38423.31 - 821.73 * highway-mpg

# 2. Model Evaluation Using Visualization

After we've built models, we need to know:

Is the model actually good?

Do we need to change it?

Is the relationship between the variables linear? Or is it more complex?


# Evaluation tools we will use:
Regression Plot

Residual Plot

Distribution Plot

#  Regression Plot (Simple Linear Regression Visualization)

**Objective:**

* Plot the actual data points (scatterplot)

* Draw the regression line drawn by the model

* See the relationship (strong/weak?, positive/negative?)

```python
import seaborn as sns
width = 12
height = 10
plt.figure(figsize=(width, height))

sns.regplot(x='highway-mpg',y='price',data=df)
plt.ylim=(0,)
plt.show()
```

# Notes:
* The line is downward (negative slope) → the relationship is negative.
* As highway-mpg increases, the price decreases.
* The points are relatively close to the line → a relatively strong relationship.

# Compare it to 'peak-rpm'

```python
plt.figure(figsize=(width, height))
sns.regplot(x="peak-rpm", y="price", data=df)
plt.show()
```

# Notes:
* The points here are more scattered around the line.

* The relationship is unclear (the price isn't constant, increasing or decreasing with peak rpm).

Conclusion: Highway mpg is a better variable for predicting price.

# Mathematical confirmation using .corr()

```python
print(df[['peak-rpm', 'price']].corr())
print(df[['highway-mpg', 'price']].corr())
```

# Correlation (r) is a number between -1 and +1:

* Close to 1 → Strong direct relationship.

* Close to -1 → Strong inverse relationship.

* Close to 0 → No clear relationship.

# Residual Plot
Objective:

We look at the residuals, which are the difference between the true value (y) and the expected value (yhat)


```python
plt.figure(figsize=(width,height))
sns.residplot(x=df['highway-mpg'],y=df['price'])
plt.show()
```

# Notes:
* If the residuals are randomly distributed around zero, the linear model is good.

* However, if there is a clear pattern (such as a curve), the linear model is insufficient.

In this example: the residuals are not uniform, the linear model is not the best choice.

# Polynomial Regression and Pipelines

Objective:
When the relationship is not linear (as we observed in residuals), we use Polynomial Regression to represent complex relationships.

Polynomial Forms:

* 2nd Order (Quadratic): Ŷ = a + b1X + b2X²

* 3rd Order (Cubic): Ŷ = a + b1X + b2X² + b3X³

* Higher Orders...


# Apply Polynomial Regression to "highway-mpg"

```python
x = df['highway-mpg']
y = df['price']
```

```python
# Calculating a Cubic Equation:
f = np.polyfit(x,y,3)
p = np.poly1d(f)
print(p)
```

```python
# Drawing the curve:
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ ' + Name)
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')
    plt.show()

PlotPolly(p, x, y, 'highway-mpg')

```

# Note:

* I notice that the curve fits the data better than the straight line.

* This means that polynomial regression provides a more accurate representation.

#  Question #4: Polynomial of 11th Order

```python
f11 = np.polyfit(x, y, 11)
p11 = np.poly1d(f11)
PlotPolly(p11, x, y, 'highway-mpg')
```

# ⚠️ Note:

Increasing the degree of the polynomial too much may lead to overfitting (the model memorizes every point, but performs poorly on new data).

# Pipeline
What is a Pipeline?

A Pipeline is a sequence of data processing and model training steps all combined in one place.

Instead of doing:

* Scaling the data

* Polynomial Transformation

* Training the model

* Then making predictions...

...each step separately in its own line,

we can perform all of them together as if they were a single step using a Pipeline.


# Pipeline building steps:

```python
# Importing the necessary tools:
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# Steps List:
# Each element is a tuple: (step name, tool we will use)
Input = [
    ('scale',StandardScaler()), # Scaling
    ('polynomial',PolynomialFeatures(include_bias=False)), # Polynomial Transformation
    ('model',LinearRegression()) # Linear Regression
]
```

```python
# Create a Pipeline:
pipe = Pipeline(Input)
```

```python
# Data preparation:
# First, we convert the data type Z to type float to avoid conversion warnings that may appear as a result of StandardScaler taking float inputs.
z = z.astype(float)
```

**Pipeline training:**

Here, the Pipeline will perform the following steps:

* Scaling (Standardizing the features)

* Polynomial Transformation (Generating polynomial features)

* Training a Linear Regression model


```python
# Pipeline training:
pipe.fit(z,y)
```

```python
# Predicting results:
ypipe = pipe.predict(z)
print(ypipe[0:4])
```

# Measures for In-Sample Evaluation
When evaluating our models, not only do we want to visualize the results, but we also want a quantitative measure to determine how accurate the model is.
# Two very important measures that are often used in Statistics to determine the accuracy of a model are:

* R^2 / R-squared

* Mean Squared Error (MSE)


| **Measure**                  | **Purpose**                                                                                         |
| ---------------------------- | --------------------------------------------------------------------------------------------------- |
| **R-squared (R²)**           | What percentage of the variation in **Y** can the model explain?                                    |
| **Mean Squared Error (MSE)** | The average of the squared differences between actual and predicted values – the lower, the better. |


#  Model 1: Simple Linear Regression (highway-mpg as X)

# Let's calculate the R^2:

```python
# lm must have been created from sklearn.linear_model.LinearRegression

# highway_mpg_fit
lm.fit(X,Y)
# Find the R^2
print(f'The R-square is: {lm.score(X,Y)}')
```

**~49.659%** : This number expresses "how much of the price change is explained by highway-mpg?"

# Let's calculate the MSE:

```python
Yhat = lm.predict(X)
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(df['price'], Yhat)
print('The mean square error of price and predicted value is: ', mse)
```

# Model 2: Multiple Linear Regression (Z as X)
#  R²:

```python
lm.fit(Z,df['price'])
print('The R-square is:', lm.score(Z,df['price']))
```

It's normal for the R² to be higher than that of the Simple Linear model because we have more features (variables) in this model.

# MSE:

```python
Y_predict_multifit = lm.predict(Z)
print('The mean square error of price and predicted value using multifit is: ', \
      mean_squared_error(df['price'], Y_predict_multifit))
```

#  Model 3: Polynomial Regression (highway-mpg with degree 3 polynomial)

# R²:

```python
from sklearn.metrics import r2_score

r_squared = r2_score(y,p(x))
print('The R-square value is: ', r_squared)
```


Here we use r2_score because we used Numpy Polynomial Model.

#  MSE:

```python
mse_poly = mean_squared_error(df['price'], p(x))
print('The mean square error of price and predicted value using polynomial fit is: ', mse_poly)
```

# Notes:

* The closer R² is to 1 → the better the model.

* The smaller the MSE → the more accurate the model.

But! Increasing the number of features or the degree of the polynomial may lead to overfitting.
